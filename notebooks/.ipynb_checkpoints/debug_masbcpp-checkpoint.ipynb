```python
#import packages
import trimesh
import numpy as np
```

```python
#set shape length, mesh size, and initial radius values
L = 5
M = 5/10
R = 1
```

```python
#CUSTOM FILE NAME!
#mesh an stl file and load it
mesh = trimesh.load_mesh('./example-stls/plate-5x5x0.2.stl')
mesh
```

```python
#check mesh faces
mesh.faces.shape
```

```python
#SET MESH DETAIL VALUE ACCORDING TO SHAPE!
#create finer mesh data and load it
new_mesh = trimesh.Trimesh(*trimesh.remesh.subdivide_to_size(mesh.vertices, mesh.faces, M))
new_mesh
```

```python
#save new mesh vertices and normals to masbcpp input coordinates and normals
np.save('./testing/input/coords.npy', new_mesh.vertices)
vn = np.array(new_mesh.vertex_normals, order = 'C')#change memory loading to column-wise for normals.npy fortran = false
np.save('./testing/input/normals.npy', vn)
```
